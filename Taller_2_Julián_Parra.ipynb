### Ejercicio 1
Escriba un programa que imprima una lista de números del 2 al 20, y que muestra si el número es divisible por 2, por 3, por ambos o por ninguno. La salida en pantalla se puede ver algo así
```
Num   Div by 2 and/or 3?
---   ------------------
2            by 2
3            by 3
4            by 2
5            neither
6            both
7            neither
```

````python
# div2_3.py
# programa imprime una lista de números del 2 al 20, 
# y muestre si el número es divisible por 2, por 3, 
# por ambos o por ninguno. 

#Se importa los módulos necesarios, en este caso 
#únicamente numpy

import numpy as np

#Se crea el for loop de 2 a 20 mediante "i" en np.arange(2,21),
#posterior a esto se coloca "if", "elif", y "else" para en caso
#de que: si el módulo de "i%2" e "i%2" son ambos 0, imprima both,
#si no es así, se evalúa la situación: si el número "i" es múltiplo 
#de 2 mediante "elif i%2 == 0" en donde si esto se cumple imprime:
#"by 2", el siguiente "elif" hace lo mismo pero con 3. Por último,
#con "else", si no se cumple ninguna de las condiciones anteriores 
#se imprime"neither". 

for i in np.arange(2,21):
    if i%2 == 0 and i%3 == 0:
        print("%2.0f    both"%(i))
    elif i%2 == 0: 
        print("%2.0f    by 2"%(i))
    elif i%3 == 0:
        print("%2.0f    by 3"%(i))
    else:
        print("%2.0f    neither"%(i))
````

### Ejercicio 2

Escriba un programa que repetidamente le pida al usuario tres valores (reales) `a`, `b`, `c`, para la ecuación 
cuadrática
$$
a*x^2 + b*x + c = 0,
$$
Usando la famosa formula para las raíces de este problema, el programa debe reconocer si 
hay raíces reales y calcular estos valores. Como resultado, muestre el número de raíces reales y sus valores. 
El programa debe parar si todos los valores son 0.  


````python
# quad_raiz.py
# Usando la formula cuadrática para despejar x
# Entradas: a, b, c
# Salidas:  x1 y x2, los dos posibles valores
# La formula es a*x**2+b*x+c = 0

#Se importa el módulo a utilizar

import math

#Se establece el while loop para que pieda los valores a,b, y c

while (True):
    tx = input("escriba los 3 números reales, si quiere terminar el programa escriba 3 0's:")
    a,b,c = tx.split()
    a = float(a)
    b = float(b)
    c = float(c)
    
#Se le dice al programa que deje de funcionar cuando se introduzcan 3 "0"s 
    
    if (a == 0 and b == 0 and c == 0):
        break
        
#Se le dice al programa que en caso de que se introduzcan valores 
#de una constante o una ecuación lineal nos comente que no es 
#una ecaución cuadrática, pero aún así que nos diga el valor de x

    elif (a == 0 and b == 0 and c !=0):
        print("\nx es una constante, cuya solución es: x =",-c,"\n")
    elif (a == 0 and b != 0 and c != 0):
        s = - c / b
        print("\naunque esta no es una ecuación cuadrática, sino una lineal, la solución es:",s,"\n")
        
#En caso de que si sea una ecuación cuadrática, se procede a ver si
#lo que está dentro de la raíz es mayor a 0, en cuyo caso van a haber
#dos soluciones para x
        
    elif ((b**2-4*a*c) > 0):
        s1 = (-b + math.sqrt(b**2-4*a*c)) / 2*a
        s2 = (-b - math.sqrt(b**2-4*a*c)) / 2*a        
        print("\nla ecuación tiene 2 raices reales, por tanto tiene 2 soluciones posibles:\n", 
              "las soluciones a la ecuación %5.2f * x^2 + %5.2f * x + %5.2f = 0 son: %5.2f y %5.2f\n"%(a,b,c,s1,s2))
        
#En caso de que dentro de la raíz el resultado es 0, va a habrer una 
#única solución para la ecuación
        
    elif ((b**2-4*a*c) == 0):
        s1s = (-b / 2*a)
        print("\nla ecuación tiene una raiz real, por tanto una solución posible, la cual es %5.2f\n"%(s1s))
        
#Finalmente en caso de que lo que esté dentro de la raíz sea negativo 
#el programa nos avisa que no hay una raíz real, por lo tanto no es posible
#hacer el cálculo
        
    elif ((b**2-4*a*c) < 0):
        print("\nno hay raices reales, no se procede a hacer el cálculo\n")
    
````

### Ejercicio 3

Modifique el programa `gfc.py` para calcular el mínimo común múltiplo (least common multiple en inglés) 
de dos números enteros. 


````python
# mcm.py
# Calcule el mínimo común divisor
# de dos enteros.

#se utiliza el comando "def" para crear la función, se crean
#las variables "ce" y "c" cuya utilidad se verá más abajo

def LCM(a,b):
    ce = 0
    c = 1
    
#se utiliza el comando "if" para decirle al programa que 
#continúe si "a" y "b" (números a evaluarles el minimo común
#múltiplo ) son mayores a cero, esto debido a que c%0 es 
#indefinido y nos echaría a perder el programa, y además porque
#no es posible encontrar mínimo común múltimplo a 0 ni a 
#numeros negativos
    
    if (a > 0 and b > 0):

#Cuando se encuentre el mínimo común múltiplo, en este caso
#"c", c%a y c%b van a ser 0, y la operación va a parar, mientras
#no se satisfaga esta condición c no va a ser mínimo común 
#múltiplo y el while loop va a continuar
            
            
        while (c%a != 0 or c%b != 0):
            
#Para no utilizar tanto procesador y agilizar la función se
#establece "c", que es el candidato a ser el mínimo común
#múltiplo. "c" es un múltiplo del número más grande entre
#"a" y "b", comenzando con 1*max(a,b), siguiendo con 2*max(a,b),...
#pero reemplazando este 1,2,... por "ce", que va creciendo de a 1 
#conforme sigue el while loop

            ce += 1
            c = max(a,b) * ce

#Cuando se encuentre el mínimo común multiplo y por tanto se 
#termine el while loop anterior la función nos va a decir cual
#es este número "c"
        
        return c
        
#Acá se le pide al usuario escribir los dos números, y posterior
#a esto se convierte este string a tipo int. este proceso
#no se incluye en la función "LCM" debido a que no siempre se quiere
#sacar el mínimo común múltiplo al tipo de información string, además
#se hace un while loop para pedir repetidamente los valores de a y b
#al usuario

while(True):
    tx = input("Escriba dos números enteros positivos, ceros para terminar: ")
    a,b = tx.split()
    a = int(a)
    b = int(b)
    if (a == 0 and b == 0):
        print("gracias por utilizar el programa")
        break
    elif (a <= 0 or b <= 0):
        print("recuerde introducir números enteros positivos")
    else:
        l = LCM(a,b)
        print("el mínimo común múltiplo de",a,"y de",b,"es:",l)
````

**NOTA**: En C o Fortran cada vez que corra el programa, Ud. obtendrá siempre la misma serie de números aleatorios. Esto no es muy buena idea, pero el código para obtener series de números aleatorios distintas cada vez que se corre el programa es complicado.
````python
# rand_check.py
# generar 10000 números aleatorios entre 0 y 1. 
# Realice un test de que tan aleatorio es el generador de números, 
# contando el número de veces que el número cae dentro de 10 rangos 
# distintos entre 0 y 1.

#Se importa el módulo necesario para crear un arreglo de 0
#de 10 elementos, en donde se van a ir adicionando las veces
#que caigan en el rango

import numpy as np
cont = np.zeros(10)

#Se crea el loop para los 1000 valores aleatorios entre 0 y 1

for i in range(10000):
    a = np.random.random()

#Se crea un for loop de 10, esto para que en cada "i" se evalúa
#gracias a un "if" si el valor de "a" cae en el rango deseado.
#ésto mediante i*0.1 <= a <(i+1)*0.1

    for i in range(10):
        if ((i*0.1)<= a and  a<(i+1)*0.1):
            cont[i] += 1

#Finalmente se crea la tabla mediante un comando for e indexando
#en el vector "cont" mediante cont[i]
            
for i in range(10):
    print("%3.0f\t%5.0f\t"%(i,cont[i]))   

````

La Tierra de manera general se divide en varias *capas*, corteza, manto y núcleo (interno y externo). Si la Tierra tiene un radio de 6371 km, y lo comparamos con un balón de Baloncesto (24.26 cm de diámetro):
Ponga en una tabla y compare 
- radio del núcleo interno
- radio hasta el tope del núcleo externo
- radio hasta el `moho` (límite manto-corteza)
- Para todos los anteriores, determine su espesor.
- La atmósfera (la parte más densa) tiene 16 km de espesor. Cuál es el espesor en nuestro ejemplo?

Espesores en km para cada capa
$$ T_{crust} = 25,    $$ 
$$T_{mantle} = 2900, $$ 
$$T_{ocore}  = 2250, $$ 
$$T_{icore}  = 1196, $$


````python
# tierra_balon.py
# Dado radio de las diferentes interfases, 
# compare si la Tierra tuviera el tamaño de un 
# balón de Basketball 

#Se introduce el módulo necesario para hacer los vectores

import numpy as np

#Se crea la función "bto" para transormar cualquier medida
#de las capas de la tierra a la escala de un balón de 
#baloncesto mediante una regla de 3, el diámetro del balon
#de baloncesto es dividido en 2 para obtener el radio.

def bto(a):
    d = a*(24.26/2)/6371
    return d

#Se introducen los nombres y los grosores de cada capa, también se 
#introducen los radios de la tierra, y mediante la función creada 
#anteriormente se calcula su correspondiente escala en un 
#balón de baloncesto

nombre = np.array(["corteza       ","manto         ","núcleo externo",
                   "núcleo interno","atmósfera     ","R nucleo ex.  ","R moho        "])
Pcol = np.array([25,2900,2250,1196,16,1196+2250,1196+2250+2900])
Pbal = bto(Pcol)

#Finalmente se imprime el nombre de las columnas y mediante
#un comando for se crea la tabla

print("        capa        |Grosor tierra(Km) |Equivalente balón(cm)\n ",
      "------------------|------------------|---------------------")

for i in range(7):
    print("%20s|%18.2f|%15.4f"%(nombre[i],Pcol[i],Pbal[i]))
````

Comparando los valores obtenidos mediante esta operación, se observa que en un balón de baloncesto poco menos de la mitad de un mm correspondería a la corteza, además la atmósfera tendría un grosor de 0.3 mm, y ésta no haría parte del balón, sino que sería una pequeña capa de aire que lo rodea. El mayor grosor es del manto, seguido por el núcleo externo, el núcleo interno, la corteza, y la atmósfera.
