### Ejercicio 2
Escriba un programa en Python que imprima en la pantalla una tabla con

` x    sinh(x)    cosh(x)`

las funciones hiperbólicas del seno y coseno, para valores de X entre 0.0 a 6.0 con incrementos de 0.5. Utilice un formato para el resultado que sea presentable. No convierta `x` a radianes. 


````python
# hyp_fun.py
# Hacer una tabla con cosh, senh
# desde 0.0 hasta 6.0, cada 0.5
"""
Se importan los módulos necesarios para hacer el arreglo 
y las operaciones trigonométricas
""" 

import math                                                   
import numpy as np                                             
"""
Se crea un vector "i" para proceder a hacer el for loop
"""
i = np.arange(13)                                             
"""
Se crea el nombre de cada columna de la tabla#
"""
print("x       sinh(x)     cosh(x)")                          
"""
Se establece el for loop "y" mediante la matriz "i".En el 
cuerpo del for loop se mutiplica "y" por 0.5 para los valores
pedidos y se le asigna el nombre x, después de esto se procede 
a sacarle el sin y cos a este valor.
"""
"""    
Por último se utiliza el comando print para que vaya llenando 
la tabla a medida que va progresando el for looop
"""
for y in i:                                                  
    x = y * 0.5                                               
    sinh = math.sinh(x)                                          
    cosh = math.cosh(x)   
    print("%2.1f    %8.3f    %8.4f   " %(x,sinh,cosh))         
print("          fin")                                         
     
````

### Ejercicio 4
Genere una tabla de multiplicar para los números del 1 al 9, es decir $1\times1, 1\times2, \dots, 1\times10$, la siguiente linea $2\times1, 2\times2, \dots, 2\times10$ y así hasta el 9. 


````python
# mult_table.py
# Genere una tabla de multiplicar para los números del 1 al 9
"""
Se importa el módulo numpy para hacer los for loops
Se crean los arreglos de 1 a 9 para "i", y de 1 a 10 
para "g"
"""

import numpy as np                    
i = np.arange(1,10)                   
g = np.arange(1,10)
"""
Se crea un loop con el valor "p" mediante "i" para el 
primer valor de la operación, y se utiliza el comando 
print para darle nombre a la tabla que se creará con 
el otro for loop
"""
for p in i:                           
    print("\n tabla del",p,"\n")
    for s in g:                       
        d = p * s                      
        print(p,"x",s,"=",d,end=", ")
    print(p,"x",10,"=",p*10)
print("\nfin")


#Se crea un for loop con el valor "s" en "g" para el 
#segundo valor de la operación, así, mediante este for 
#se va multiplicando un único valor "p" por un segundo 
#valor "s" que va cambiando e imprimiendo su resultado
````

### Ejercicio 6
Genera una tabla, con los siguientes valores
$$
x \quad  \log10(x) \quad \log(x) \quad \sqrt(x)
$$
Haga tantos `loops` como sea posible, empezando con $i=0, x=1.01$. Para el siguiente `loop` 
- $i = 0, \quad x_i = 1.01 $
- $i = 1, \quad x_i = x_{i-1}(i+1)$
- $ i = 2, \quad \cdots $

En cada loop calcule los logaritmos y raíz cuadrada. Eventualmente el resultado no va a caber en el computador (los números serán muy grandes). En su equipo, hasta cuándo puede hacer esto? Esto muestra el límite de precisión del computador. 
  
Los primeros valores de $x$ deben ser $1.01$, $2.02$, $6.06$, $24.24$, $...$

````python
### Ejercicio 6
"""
Se importan los módulos necesarios para hacer el arreglo 
del for loop, y para hacer operaciones trigonométricas
"""
import numpy as np          
import math                  
"""
Se crea el arreglo "m" para 171 valores, comenzando con 0
y terminando en 170 mediante el comando arange, también
se le da el valor inicial 1.01 a "x"
"""
m = np.arange(171)        
x = 1.01                     
"""
En el siguiente paso se le da un nombre a cada columna de 
la tabla, con los espacios de acuerdo al formato que se 
utilizará en el último comando print
"""
print("  i                                                                 x  log10(x)  log(x)    "
      "                        sq(x)") 

"""
Se crea el valor "x" multiplicando el valor inicial  de "x"
(1.01) por "i + 1", al ser un comando loop, este nuevo valor
se irá modificando a medida que avanza el loop, posterior
a esto se hacen las operaciones log, log10, y raíz cuadrada
"""
"""
Por último se utiliza el comando print para llenar la tabla,
en este caso es muy útil adicionar | debido a que organiza 
la tabla (o por lo menos los primeros números) como mejor 
convenga
"""
for i in m:                 
    x    = x * (i + 1)      
    l10  = math.log10(x)
    l    = math.log(x)
    sq   = math.sqrt(x)
    print("%3.0f|%66.2f|%7.4f|%8.4f|%32.4f|"%(i,x,l10,l,sq))
````

### Ejercicio 7

Asumiendo que la velocidad absoluta de una placa tectónica es de $55$ mm/a, calcule a que distancia se encuentra una roca de $1$Ma, de $8$Ma y de $60$Ma. Si la placa está subduciendo con un ángulo de 60 grados, a qué profundidad estaría?

````python
# Dada la velocidad de un movimiento de una placa tectónica
# calcule cuanto se ha desplazado en 1, 8, 60 Ma
"""
Se importa el módulo math, necesario para el cálculo de
seno
"""
import math                         
"""
Se calcula la velocidad "v" en Km/Ma, esto se hace 
multiplicando por 1000000(años en un Ma) y dividiendo
por 1000000(mm en un Km)
"""
v = 55*1000000/1000000              
"""
Se pasa de grados a radianes el ángulo de subducción 
mediante la multiplicación de este por pi, y dividiéndolo
por 180 (regla de 3), ésto con el objetivo de calcular
el seno "sin". 
"""
sin = math.sin(60*math.pi/180)       
"""
Se adicionan los tiempos 1 Ma, 8 Ma, y 60 Ma, nombrándolos
"a, b, y c" respectivamente
"""
a = 1                                
b = 8                                
c = 60                                 
"""
Se calcula la distancia mediante velocidad "v" y tiempo "a,b
y c" para los 3 casos
"""
da = v * a                          
db = v * b                      
dc = v * c
"""
Se utiliza cateto opuesto = sin x hipotenusa (distancias
calculadas anteriormente) para encontrar la profundidad
"""
ha = da * sin                       
hb = db * sin                        
hc = dc * sin
"""
Por último se utiliza el comando print para presentar
los resultados
"""
print("para una velocidad absoluta de 55 mm/año a lo largo de 1 Ma, 8 Ma, y 60 Ma, se observan profundidades de subducción"
     "n de: %7.2fKm, %7.2fKm, y %7.2fKm respectivamente."%(ha,hb,hc),"\n\nEn caso de que no sea un margen destructivo"
      " las distancias que recorren las placas son: %7.2fKm, %7.2fKm, y %7.2fKm respectivamente"%(da,db,dc))
````
